```python
from benchmark import benchmark
import matplotlib.pyplot as plt
```

```python
n_values, avg_times_without_cache, avg_times_with_cache = benchmark(1000)
plt.figure(figsize=(10, 6))
plt.plot(n_values, avg_times_without_cache, label='Without Cache', marker='o')
plt.plot(n_values, avg_times_with_cache, label='With Cache', marker='s')
plt.xlabel('Number of Iterations (n)')
plt.ylabel('Average Execution Time (seconds)')
plt.title('Average Execution Time by Number of Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
